# 卷积神经网络 CNN

# CNN原理

1 局部性
对一张图片而言，需要检测图片中的特征来确定图片的类别，通常情况下这些特征是图片的一些局部区域决定的


2 相同性 可以用同样的检测模式去检测不同图片的相同特征（只不过这些特征在图片的不同位置），但是我们所要做的检测操作是一样的

3 不变性 对一张大图片进行subsampling，图片的基本性质不变

# CNN组成


CNN由卷积层，池化层和全连接层组成，通过堆叠这三种层次结构，形成一个完整的卷积神经网络（卷积+池化+卷积+池化+。。。+全连接）
卷积层和全连接层具有参数，而激活层和池化层则没有（参数也是通过梯度下降来更新的）

# 卷积层


由一些可学习的滤波器组成

每个滤波器在空间上（宽度&高度）上都比较小，但深度与输入数据一致（比如说，输入是RGB三通道的，那么滤波器也是三通道的）

多个小滤波器比一个大滤波器好

在前向传播的时候，让每个滤波器在输入数据的宽度和高度上滑动，然后计算整个滤波器和输入数据任意一处的内积

计算结果是一个二维的激活图，表示了原图片对于滤波器的反应（【学习到特定视觉特征时，激活】边界、颜色、轮廓等等）

每个卷积层上会有一整套集合的滤波器（比如20个），那么就会形成20张二维的不同激活图，这些激活图在深度方向上叠加起来就形成了卷积层的输出


卷积层的局部连接

利用图片特征的局部性，只需要通过局部就可以提取出相应的特征

这就是神经元的感受野（receptive field）（也就是滤波器的宽和高）



比如感受野是5*5*3，那么卷积层的每个神经元就有5*5*3=75个权重


卷积层的输出深度：和使用的滤波器数量一致（每一种滤波器就是在输入数据中寻找一种特征）

步长：每一个滤波器每一次移动几个像素点（步长大的话能够使得输出的数据在空间上更小）



边界填充——将输入数据用0在边界进行填充——可以让输入和输出在空间上尺寸一致



第一个-2就是0*1+1*0+2*-1=-2

输出的尺寸=（数据长/宽-卷积层长/宽的感受野尺寸+2*边界填充数量）÷步长+1
