# 数据处理


中心化

这是一种最常见的处理方法，每个特征维度都减去相应的均值，来实现中心化
这样使得数据变成0均值

标准化

数据变成0均值后，可以使用标准化，来让数据不同特征维度都有着相同的规模

一种方法是每个元素都除以标准差，这样新数据的分布就接近于高斯分布

另一种方法是让每个特征维度的最大最小值按照比例缩放到-1~1



PCA 主成分分析
改变数据的特征向量，降低数据的维度

白噪声
PCA 的基础上将各个维度转化成0均值，协方差矩阵为1的多元高斯分布


# 权重初始化

全0 初始化

最直观，但最不应该被采用
将参数全部初始化成零

这个的问题是，每个神经网络中的神经元权重相同，那么每个神经元都会计算出相同的结果，反向传播的时候会计算出相同的梯度，最后导致所有权重有相同的更新

随机初始化

首先权重初始化的时候尽量接近0，但不能等于0
所以可以初始化权重为一些靠近0的随机数(eg，高斯随机化，均匀随机化

稀疏初始化

先全0初始化，然后随机挑选几个参数附上一些随机值（实际应用中较少）

# 防止过拟合

如果网络容量过大，可能导致过拟合，那如何防止呢

正则化

L2正则化，对权重过大的部分进行惩罚
直接在损失函数中增加0.5入w^2，入是正则化强度
加0.5的原因是这样的话变成梯度的时候就是入w
这样可以让参数更新之后更接近0

L1正则化
在损失函数中增加权重的1范数，入|w|


最大范数限制，|w|有一个上限c，这样可以让网络不会太爆炸，因为更新总是有限的

实际过程中，通常选择L2正则化

Dropout
在训练网络的时候以概率P来保留每一个神经元（每一次训练的时候有一些神经元会被设置成0）


但是在预测的时候，我们不用这样处理，预测的时候保留整个网络，但是对每一层网络的输出应用P的缩放，这样能够让预测的行为和训练的行为相同

另一种理解dropout的思路是把他看成几个minibatch的集成


实际应用中，一般采用全局权重的L2正则化，配以dropout来防止过拟合
