# torch.nn.init 初始化

```python
import torch
print(torch.nn.init.uniform_(torch.Tensor([[1,2],[3,4]]),a=0,b=100))
#对第一个参数这个tensor进行初始化，在0~100上均匀分布
#（如果不赋值的话,a默认为0，b默认为1）
```

```python
print(torch.nn.init.normal_(torch.Tensor([[1,2],[3,4]]),mean=0,std=100))
#对第一个参数这个tensor进行初始化，以mean为均值，std为标准差
#（如果不赋值的话,mean默认为0，std默认为1）
```

```python
print(torch.nn.init.constant_(torch.Tensor([[1,2],[3,4]]),0))
#对第一个参数这个tensor进行初始化，初始化为一个常数0
```

xavier初始化——输入和输出的方差相同

pytorch提供了两种xavier初始化
一种忘均匀分布走U(-a,a)
参数里面的gain默认为1

fan_in是输入的神经元个数
fan_out是输出的神经元个数

a的计算公式


```python
print(torch.nn.init.xavier_uniform_(torch.Tensor([[1,2],[3,4]]),gain=10))
#对第一个参数这个tensor进行初始化，初始化为一个常数0
```

另一种是正态分布N(0,std)


Xavier初始化后的参数在tanh中表现很好，但在relu中表现的很差

```python
print(torch.nn.init.xavier_normal_(torch.Tensor([[1,2],[3,4]]),gain=10))
#对第一个参数这个tensor进行初始化，初始化为一个常数0
```

kaiming初始化

xavier的改良版，在relu上表现很好

也是由均匀分布和正态分布两个

均匀分布

a默认为0

正态分布，均值为0，方差


参数是tensor，a，mode
mode默认是fan_in，也可以是fan_out
fan_in保持weight的方差在前向传播中不变，fan_out保证weight的方差在反向传播中不变


基本上基于relu的激活函数就是使用he initialization

```python
print(torch.nn.init.kaiming_normal_(torch.Tensor([[1,2],[3,4]]),a=1,mode='fan_in'))
```

```python
print(torch.nn.init.kaiming_uniform_(torch.Tensor([[1,2],[3,4]]),a=1,mode='fan_out'))
```
