# various  forms of gradient descent

# SGD

这是梯度下降法的一个小变形，每一次使用一个batch的数据进行梯度计算（而不是全部数据）

这是因为现在深度学习的数据量很大，所以每次计算所有的梯度是不现实的（因为这样会导致运算时间非常的长）【同时每次计算的梯度都差不多，失去了一些随机性，可能会陷入局部误差】

使用SGD 反而可能跳出局部最小点

# Momentum

在随机梯度下降的同时，增加动量（momentum)


这样的方法会基于上一次的梯度，也就是说参数的更新不仅仅取决于当前的梯度，而且也取决于之前的梯度


△L是梯度，θ是移动方向，虚线的部分就是动量，也就是之前一次移动方向的负方向

移动方向是之前一次移动方向的负方向（也就是动量）和梯度的矢量和

当然，对于动量来说还有一个变形，也就是nesterov，这个方法是去计算经过动量更新之后位置的梯度

# adagrad

一种自适应学习的方法（adaptive）


学习率在不断地减少
这个根号很重要，因为没有根号的话，算法的表现会非常的差
c是一个平滑参数，防止分母为0

但adagrad的缺点是会导致学习过早停止（因为学习率降到0了嘛）

# RMSprop 

adagrad的一种改进



相比于adagrad来说，这里多了一个α，说明RMSprop不是仅仅将前面所有的梯度平方求和，而是增加了一个衰减率

这样可以有效地缓解过早结束学习的问题

# adam

可以想成 RMSprop+momentum
