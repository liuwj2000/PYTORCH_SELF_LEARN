# 神经网络的一些概念

#激活函数 

```python
#sigmoid
#sigmoid的数学表达式是1/(1+e^-x)
import numpy as np
import matplotlib.pyplot as plt

a=np.linspace(-10,10,100)
b=1/(1+np.exp(-a))

plt.plot(a,b)
ax=plt.gca()
ax.spines['right'].set_position(('data',0))
ax.spines['left'].set_position(('data',0))
ax.spines['top'].set_position(('data',0))
ax.spines['bottom'].set_position(('data',0))
plt.show()
```

sigmoid的数学表达式是1/(1+e^-x),图形如上
这个激活函数将实数转化成0~1

但sigmoid存在很多问题
(1) 会造成梯度损失

```python
#tanh
#性能上比sigmoid好（因为采用了零均值），但是还是存在梯度消失的问题
c=np.tanh(a)
plt.plot(a,c)
ax=plt.gca()
ax.spines['right'].set_position(('data',0))
ax.spines['left'].set_position(('data',0))
ax.spines['top'].set_position(('data',0))
ax.spines['bottom'].set_position(('data',0))
plt.show()
```

```python
#relu
#函数表达式 f(x)=max(0,x)
d=[]
for i in a:
    if i<0:
        d.append(0)
    else:
        d.append(i)
d=np.array(d)
plt.plot(a,d)
ax=plt.gca()
ax.spines['right'].set_position(('data',0))
ax.spines['left'].set_position(('data',0))
ax.spines['top'].set_position(('data',0))
ax.spines['bottom'].set_position(('data',0))
plt.show()
#梯度线性，所以不存在梯度消失的问题
```

```python
#leaky relu
#负数部分给一个很小的常数，小梯度（也可以通过学习学出这个梯度参数）
e=[]
for i in a:
    if i<0:
        e.append(0.05*i)
    else:
        e.append(i)
e=np.array(e)
plt.plot(a,e)
ax=plt.gca()
ax.spines['right'].set_position(('data',0))
ax.spines['left'].set_position(('data',0))
ax.spines['top'].set_position(('data',0))
ax.spines['bottom'].set_position(('data',0))
plt.show()
```

maxout
max(w1x+b1,w2x+b2)
也就是说，relu只是w1=0,b1=0,w2=1,b2=0这样的特殊maxout

一般来说，N层神经网络不会把输入层算进去，因此一个一层的神经网络是指只有输入层和输出层的神经网络
同时，一般输出层是没有激活函数的

增大网络的层数和每层的节点数，意味着增大网络的容量。容量的增加意味着网络有更好的表现能力


这个图是对三个网络模型做二分类的结果
每个网络模型都是一个隐藏层，但是每个隐藏层的节点数量不一样（3，6，20个）
可以看出隐藏结点越多模型越复杂，但是太过的结点会导致过拟合

但是，并不是容量小的模型就好，因为这样的模型在使用梯度下降的时候会很困难
因为神经网络的损失函数一般是非凸的，容量小的网络更容易陷入局部极小点

所以实际上我们还是使用大容量的网络去训练，同时使用一些方法来控制网络的过拟合
