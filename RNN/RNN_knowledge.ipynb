# 循环神经网络RNN


卷积神经网络相当于人的视觉，循环神经网络相当于人的记忆

将网络的输出保存在一个记忆单元中，这个记忆单元和下一次的输入一起进入神经网络



双向循环神经网路：网络先从序列的正方向读取数据，再从反方向读取数据，最后将两种结果结合在一起

# LSTM——RNN变式

long short term memory nerworks

LSRM的抽象网络结构：


LSTM由三个门控制——输入门、遗忘门（记忆单元）、输出门

由于遗忘门的作用，LSTM具有了长时记忆的功能，对于给定的任务，遗忘门可以自己学习保留多少之前的记忆

将t时刻之间的以及乘上一个0~1的遗忘程度进行衰减，然后加上t时刻的输入，得到t时刻学到的记忆，更新后传出网络，作为t+1时刻的记忆单元



t-1时刻网络记忆在t时刻的衰减系数，是通过t时刻网络的输入&t-1时刻网络的输出得到的

 sigmoid激活函数

t时刻学到的记忆


首先也计算一个衰减系数（也是用前面一样的方法学习一个系数）

同时计算当前状态学习到的记忆


两个一结合，就得到了t时刻下学习到的记忆Ct

上面是输入门和遗忘门，下面介绍输出门

时刻t的网络输出ht取决于t时刻的记忆状态和t时刻的输入xt，t-1时刻的网络输出h（t-1）
